```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Student_Placement_Career_Prediction_Dataset.csv')
df.head()
df.info()
```

```python
df.isna().sum()
df.duplicated().sum()
```

```python
df['student_id'] = df['student_id'].fillna("Not A Student")
df['age'] = df['age'].fillna("Missing")
df['gender'] = df['gender'].fillna("Male or Female")
df['branch'] = df['branch'].fillna("IT, CSE, Civil, EEE, or Mechanical")
df['cgpa'] = df['cgpa'].fillna("Missing")
df['intership_count'] = df['internship_count'].fillna("Missing")
df['project_count'] = df['project_count'].fillna("Missing")
df['certifications_count'] = df['certifications_count'].fillna("Missing")
df['coding_skills_score'] = df['coding_skills_score'].fillna("Missing")
df['communication_skills_score'] = df['communication_skills_score'].fillna("Missing")
df['soft_skills_score'] = df['soft_skills_score'].fillna("Missing")
df['hackathon_participation'] = df['hackathon_participation'].fillna("Yes or No")
df['placement_status'] = df['placement_status'].fillna("Placed or Not Placed")
```

```python
df = df.drop_duplicates()
```

```python
df.describe()
```

```python
df['age'].quantile([0.1,0.25,0.5,0.75])
```

```python
Q1  = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['age'] < lower_bound) | (df['age'] > upper_bound)]
print(f"\nColumn: {'age'}")
print(f"Number of outliers: {outliers.shape[0]}")
```
